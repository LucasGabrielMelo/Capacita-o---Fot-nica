# Phase and group speed #

This code has the purpose to view the phase speed and the group speed of the propaganting modes in a waveguide with cutoff frequency $f_{c}$ as a function of the frequency $f$. These speeds are giveng by: 

$$v_{p} = \frac{c}{\sqrt{1-\large{\left(\frac{\omega_{c}}{\omega}\right)^{2}}}}$$

$$v_{g} = c\sqrt{1-\left(\frac{\omega_{c}}{\omega}\right)^{2}}$$

Where $c$ is the lightspeed, $\omega_{c} = 2\pi f_{c}$ and $\omega = 2\pi f$.

```python
import numpy as np
import matplotlib.pyplot as plt

c = 3 * 1e8 # lightspeed in m/s
fc =  10 * 1e12 # cutoff frequency in Hz
wc = 2 * np.pi * fc # cutoff angular frequency in rad/s

f = np.linspace(1, 50 * 1e12, 1000) # frequency range in Hz
w = 2 * np.pi * f # angular frequency range in rad/s

vp = c / np.sqrt(1 - (wc / w)**2) # phase speed
vg = c * np.sqrt(1 - (wc / w)**2) # group speed

plt.figure(figsize=(8,6))
plt.plot(f / 1e12, vp / 1e3, label='phase speed', color = 'r')
plt.plot(f / 1e12, vg / 1e3, label='group speed', color = 'b')
plt.xlabel(r'frequency ($THz$)')
plt.ylabel(r'speed ($km/s$)')
plt.yticks(np.arange(0,2,0.2)*1e6)
plt.ylim(0,1*1e6)
plt.xlim(0,50*1e12/1e12)
plt.axhline(c / 1e3, color = 'black', linestyle = '--', label = 'lightspeed')
plt.legend()
plt.grid(True)
plt.show()
```
